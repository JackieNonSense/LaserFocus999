```python

import sys
import os
from ultralytics import YOLO


DATA_YAML_FILE = '/content/archive/data.yaml'

MODEL_NAME = 'yolo11s.pt'
final_model_path = '/content/insect_detection_project/yolov11s_run/weights/best.pt'

print(f"model {final_model_path}")

best_model = YOLO(final_model_path)

metrics = best_model.val(
data=DATA_YAML_FILE,
    split='test',
    save_txt=True,
    save_conf=True,
    save_crop=True,
    plots=True
)

print(f"mAP50 (IoU=0.5):        {metrics.box.map50:.4f}")
print(f"mAP50-95 :    {metrics.box.map:.4f}")
print(f"Mean Precision (test): {metrics.box.mp:.4f}")
print(f"Mean Recall (test):    {metrics.box.mr:.4f}")

ap50_scores = metrics.box.all_ap[:, 0]
class_indices = metrics.box.ap_class_index

for i, class_idx in enumerate(class_indices):
    class_name = metrics.names.get(class_idx, f"Class {class_idx}")
    ap_value = ap50_scores[i]
    print(f"  {class_name.ljust(15)}: {ap_value:.4f}")

```
